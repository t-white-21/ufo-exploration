```python
import pandas as pd
from geopy.geocoders import Nominatim
import folium
```

```python
usa_sites = pd.read_csv('../data/usa_ufos_clean.csv', nrows = 132308)
```

```python
#https://www.youtube.com/watch?v=j81vqBf5ysE
geolocator = Nominatim(user_agent='my_geocoder')
```

```python
def geocode_address (row):
    location = geolocator.geocode(f"{row['city']}, {row['state']}, {row['country']}")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
    
```

```python
usa_sites['Latitude'], usa_sites['Longitude'] = zip(*usa_sites.apply(geocode_address, axis=1))

#Calculate mean of coordinates for map center
map_center = [usa_sites['Latitude'].mean(), usa_sites['Longitude'].mean()]

#make map with folium
mymap = folium.Map(location=map_center, zoom_start=4) #adjust as needed

#Add markers
for index, row in usa_sites.iterrows():
    if row['Latitude'] and row['Longitude']:
        folium.Marker([row['Latitude'], row['Longitude']], popup=row['City']).add_to(mymap)

#Save

mymap.save("geocoded_map.html")
```
