```python
import pandas as pd
from bs4 import BeautifulSoup as BS
import re
```

```python
import requests

URL = "https://nuforc.org/subndx/?id=e202312"
page = requests.get(URL)
soup = BS(page.content, "lxml")

print(page.text)
```

```python
results = soup.find('tbody')
results
```

```python
results.find_all('tr')
```

```python
rows = results.find_all('tr')
```

```python
occured = []
city = []
state = []
country = []
shape = []
summary = []
reported = []
posted = []
image = []

for x in range(0,100):
      o = rows[x].find_all('td', style='')[1].text.strip()
      c = rows[x].find_all('td', style='')[2].text.strip()
      s = rows[x].find_all('td', style='')[3].text.strip()
      co = rows[x].find_all('td', style='')[4].text.strip()
      sh = rows[x].find_all('td', style='')[5].text.strip()
      sum = rows[x].find_all('td', style='')[6].text.strip()
      r = rows[x].find_all('td', style='')[7].text.strip()
      p = rows[x].find_all('td', style='')[8].text.strip()
      i = rows[x].find_all('td', style='')[9].text.strip()
   
      occured.append(o)
      city.append(c)
      state.append(s)
      country.append(co)
      shape.append(sh)
      summary.append(sum)
      reported.append(r)
      posted.append(p)
      image.append(i)



print(country)

```

```python
ufo_df = pd.DataFrame(list(zip(occured, city, state, country, shape, summary, reported, posted, image)), columns=['occured', 'city', 'state', 'country', 'shape', 'summary', 'reported', 'posted', 'image'])
```
